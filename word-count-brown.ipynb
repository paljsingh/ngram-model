```python
"""
Program to find count of all words in a corpus and 
print words having a minimum threshold of occurance
"""
from collections import defaultdict

class WordCounter:
    
    def __init__(self):
        self.words = defaultdict(lambda: 0)
        self.word_tokenizer = RegexpTokenizer(r'\w+')
    
    def count_words(self):
        for file in nltk.corpus.brown.fileids():
            for word in nltk.corpus.brown.words(file):
                # tokenize, remove punctuations.
                word_token = self.word_tokenizer.tokenize(word)
                if word_token:
                    # convert all tokens to lowercase and count frequency
                    self.words[word_token[0].lower()] += 1

    def get_words(self, min_frequency=1):
        word_dict = dict()
        for word, frequency in self.words.items():
            if frequency >= min_frequency:
                word_dict[word] = frequency
        return word_dict
                    
                    
if __name__ == '__main__':
    w = WordCounter()
    w.count_words()
    
    # print sorted output
    print(sorted(w.get_words(min_frequency=3).items(), key=lambda x: x[1], reverse=True))

    
```
